```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# read files using pandas and transform them into dataframes
df1 = pd.read_csv('../price-and-sales-data.csv')
df1.drop_duplicates() # remove duplicate rows (if any)
print(df1.iloc[0]) # display the first row of the dataframe
print(df1.shape[1]) # display the number of columns in the dataframe
df1.head()
```

```python
# renaming few columns
df1 = df1.rename(columns={'4046':'code1', '4225':'code2', '4770':'code3'})
```

Data exploration and cleaning

```python
df1.describe(include='all') # this will provide a statistics of all the fields of a dataframe.
```

```python
df1 = df1.dropna()
df1.shape
```

```python
df1.pivot_table(index='region', values='TotalVolume', margins=True)
```

```python
# Extract month from date, so that we can see the monthly avocado sale trend
df1['Date'] = pd.to_datetime(df1['Date'])
df1['month']  = df1['Date'].dt.month
df1.head()
```

# Data Visualization

```python
# first, let's check the Average Price distribution
plt.figure(figsize = (10,8))
plt.title('Average price distribution of avocados')
sns.distplot(df1['AveragePrice'])
```

We can see here the average price ranges of avocadoes are approximatelly in between 0.3 and 3.4. The data is more concentrated approximately between prices 1.0 and 1.9 

There are two categories of avocadoes in the data i.e. conventional and organic. Let's visualize the statistics of both types.

```python
plt.title('Avocados count on categories')
sns.countplot('type', data = df1)
```

By above plot, we can see both categories are sold almost equally.
Now we should also check the price range of both types of avocados.

```python
plt.figure(figsize = (10,8))
plt.title('Average price range')
sns.boxplot(x='type', y='AveragePrice', data=df1)
```

We can see the organic avocados are significatly expensive than the conventional avocados. 

Let's check the region-wise avocados prices and volumes. By which we can identify the region where avocados are most expensive and the region where avocados consumption is maximum.

```python
plt.figure(figsize = (14,8))
plt.title('Average price range in area')
plt.xticks(rotation = 'vertical')
sns.boxplot(x='region', y='AveragePrice', data=df1)
```

By above plot we can see the avocados are most expensive in region San Francisco and HartforSpringField. Whereas prices are relatively cheap in SouthCentral, DallasFtWorth and Houston is the cheapest.

```python
plt.figure(figsize = (14,8))
plt.title('Organic and conventional avocados average price range in area')
plt.xticks(rotation = 'vertical')
sns.boxplot(x='region', y='AveragePrice', hue='type', data=df1)
```

Most expensive organic avocados are sold in San Francisco region.

Now let's take a look to yearly AveragePrice of both types of avocados

```python
plt.figure(figsize = (14,8))
plt.title('Yearly Averge Price')
sns.lineplot(x='year', y='AveragePrice', data=df1, hue='type')
```

From the above plot we can see the avocados are started getting expensive from mid of 2016 and the peak prices are at the start of 2017.

Check for monthly Average Price

```python
plt.figure(figsize = (14,8))
plt.title('Monthly Averge Price')
sns.lineplot(x='month', y='AveragePrice', data=df1, hue='type')
```

We can see here the avocados are most expensive in the month of September and October. Hence, there is a seasonal trend in the data.
